```python
import cv2
from ultralytics import YOLO

# Load your trained YOLOv10 food detection model
model = YOLO("trained_yolov10_food_model.pt")

# Initialize the webcam
cap = cv2.VideoCapture(0)  # Use 0 for the default webcam

# Loop to capture frames from the webcam
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture frame from webcam. Exiting...")
        break

    # Use the YOLO model to make predictions on the current frame
    results = model(frame)

    # Loop through each detection result and draw bounding boxes and labels
    for result in results:
        for box in result.boxes:
            # Extract bounding box coordinates
            x1, y1, x2, y2 = map(int, box.xyxy[0])  # Convert to integers
            cls = int(box.cls[0])  # Class ID
            conf = float(box.conf[0])  # Confidence score

            # Draw the bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green rectangle

            # Display the label and confidence score
            label = f"{model.names[cls]}: {conf:.2f}"
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Show the frame with the detected food items
    cv2.imshow("YOLOv10 - Food Detection", frame)

    # Press 'q' to exit the webcam feed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
